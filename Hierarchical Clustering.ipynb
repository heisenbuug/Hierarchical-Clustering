```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
D = 2
S = 4
matrix1 = np.array([0, 0])
matrix2 = np.array([S, S])
matrix3 = np.array([0, S])

N = 900
X = np.zeros((N, D))
X[:300, :] = np.random.randn(300, D) + matrix1
X[300:600, :] = np.random.randn(300, D) + matrix2
X[600:, :] = np.random.randn(300, D) + matrix3
```

```python
Z = linkage(X, 'ward')
print(Z.shape)
plt.title("Ward")
dendrogram(Z)
plt.show()
```

```python
Z = linkage(X, 'single')
print(Z.shape)
plt.title("Single Linkage")
dendrogram(Z)
plt.show()
```

```python
Z = linkage(X, 'complete')
print(Z.shape)
plt.title("Complete Linkage")
dendrogram(Z)
plt.show()
```
